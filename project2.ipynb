```python
cus_mon = df.groupby(['month'],as_index=False)['Customer ID'].count().sort_values(by='Customer ID',ascending=False)
cus_mon
```

```python
#gender-wise sales 
gender = df.groupby(['month'],as_index = False)["Gender"].value_counts()
gender
```

```python
grp2 = sns.barplot(x='month',y='count', hue='Gender',data= gender)
for bars in grp2.containers:
    grp2.bar_label(bars)
plt.title("Monthly sales analysis based on gender")
plt.legend(
    loc='upper left',
    bbox_to_anchor=(1,1))
plt.show()
```

##  **How do customers adapt their shopping habits during seasonal trends?**

```python
df['season'] = df['month'].apply(lambda x : 'Winter' if x in [12,1,2] else ('Spring' if x in [3,4,5] else ('summer' if x in [6,7,8]else'Fall')))
```

```python
plt.figure(figsize=(12,6))
grp = sns.barplot(x='season',y='Total Amount',data =df, ci =None)
for bars in grp.containers:
    grp.bar_label(bars)
plt.title('Total Sales by Season')
plt.xlabel('Season')
plt.ylabel('Total Amount')
plt.show()
```

```python
# sales by season and gender 
plt.figure(figsize=(12,6))
sns.barplot(x='season',y='Total Amount',hue= 'Gender',ci =None,data = df)
plt.title('Sales by Season and Gender')
plt.xlabel('Season')
plt.ylabel('Total Amount')
plt.show()
```

```python
# sales by season
plt.figure(figsize=(12,6))
grp=sns.barplot(x='month', y='Total Amount',hue='Gender', data=df,ci=None)
for bars in grp.containers: 
    grp.bar_label(bars)
plt.title('Total Sales by Season')
plt.xlabel('Season')
plt.ylabel('Total Amount')
plt.show()
```
